## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Separate the two output features
y_train_output1 = y_train[:, 0]  # First column for the output ranging from 0.1 to 1.0
y_train_output2 = y_train[:, 1]  # Second column for binary classification

y_test_output1 = y_test[:, 0]  # First column for the output ranging from 0.1 to 1.0
y_test_output2 = y_test[:, 1]  # Second column for binary classification
```

```python
print(y_train_output2)
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize input features
scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)

```

```python
print(X_train_normalized)
```

```python
print(type(y_train_output1))
print(type(y_train_output2))
print(type(y_test_output1))
print(type(y_test_output2))
```

## Building Model

```python
input_shape = (49,)
inputs = tf.keras.Input(shape=input_shape)
x = tf.keras.layers.Dense(units=64, activation='relu')(inputs)
x = tf.keras.layers.Dropout(0.2)(x)
x = tf.keras.layers.Dense(units=64, activation='relu')(x)
x = tf.keras.layers.Dense(units=32, activation='relu')(x)
output_1 = tf.keras.layers.Dense(units=1, activation='linear', name='output_1')(x)  # Output 1: Regression
output_2 = tf.keras.layers.Dense(units=1, activation='sigmoid', name='output_2')(x)  # Output 2: Classification

model = tf.keras.Model(inputs=inputs, outputs=[output_1, output_2])

# Check the model output names
print("Model Output Names:", model.output_names)  
```

```python
model.compile(
    loss={
        'output_1': 'mean_squared_error',  # Regression loss function
        'output_2': 'binary_crossentropy'  # Binary classification loss function
    },
    optimizer='adam',
    metrics={
        'output_1': ['mae', 'mse'],
        'output_2': 'accuracy'
    }
)

model.fit(X_train_normalized, {'output_1': y_train_output1, 'output_2': y_train_output2},
          validation_data=(X_test_normalized, {'output_1': y_test_output1, 'output_2': y_test_output2}),
          epochs=50, batch_size=32)

```

```python
# Predict on the test data
y_pred_output1, y_pred_output2 = model.predict(X_test_normalized)

# For output_1 (Regression)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae_output1 = mean_absolute_error(y_test_output1, y_pred_output1)
mse_output1 = mean_squared_error(y_test_output1, y_pred_output1)
r2_output1 = r2_score(y_test_output1, y_pred_output1)

print("Output 1 - Mean Absolute Error:", mae_output1)
print("Output 1 - Mean Squared Error:", mse_output1)
print("Output 1 - R^2 Score:", r2_output1)



# For output_2 (Classification)
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Convert probabilities to binary predictions (0 or 1) for classification
y_pred_output2_binary = (y_pred_output2 > 0.5).astype(int)

accuracy_output2 = accuracy_score(y_test_output2, y_pred_output2_binary)
conf_matrix_output2 = confusion_matrix(y_test_output2, y_pred_output2_binary)
classification_rep_output2 = classification_report(y_test_output2, y_pred_output2_binary)

print("Output 2 - Accuracy:", accuracy_output2)
print("Output 2 - Confusion Matrix:")
print(conf_matrix_output2)
print("Output 2 - Classification Report:")
print(classification_rep_output2)
```

## K-fold Cross Validation

```python
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


def gelu_activation(x):
    return 0.5 * x * (1 + tf.keras.activations.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * tf.keras.backend.pow(x, 3))))



num_folds = 5

# Initialize lists to store the evaluation metrics for each fold
mse_output1_list = []
accuracy_output2_list = []

# Create the KFold object
kfold = KFold(n_splits=num_folds, shuffle=True, random_state=42)


# Create the model
input_shape = (49,)
inputs = tf.keras.Input(shape=input_shape)
x = tf.keras.layers.Dense(units=256, activation='relu')(inputs)  # GELU activation
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dropout(0.2)(x)
x = tf.keras.layers.Dense(units=256, activation='relu')(x)  # GELU activation
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dropout(0.2)(x)
x = tf.keras.layers.Dense(units=256, activation='relu')(x)  # GELU activation
x = tf.keras.layers.BatchNormalization()(x)
x = tf.keras.layers.Dropout(0.2)(x)
x = tf.keras.layers.Dense(units=128, activation='relu')(x)  # GELU activation
   

output_1 = tf.keras.layers.Dense(units=1, activation='linear', name='output_1')(x)  # Output 1: Regression
output_2 = tf.keras.layers.Dense(units=1, activation='sigmoid', name='output_2')(x)  # Output 2: Classification

model = tf.keras.Model(inputs=inputs, outputs=[output_1, output_2])

# Perform 5-fold cross-validation
for fold, (train_indices, val_indices) in enumerate(kfold.split(X_train_normalized)):
    print(f"Fold {fold+1}/{num_folds}")
    
    # Split the data into training and validation sets for this fold
    X_train_fold = X_train_normalized[train_indices]
    X_val_fold = X_train_normalized[val_indices]
    y_train_output1_fold = y_train_output1[train_indices]
    y_train_output2_fold = y_train_output2[train_indices]
    y_val_output1_fold = y_train_output1[val_indices]
    y_val_output2_fold = y_train_output2[val_indices]
    
    
    # Compile the model
    model.compile(
        loss={
            'output_1': 'mean_squared_error',  # Regression loss function
            'output_2': 'binary_crossentropy'  # Binary classification loss function
        },
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        metrics={
            'output_1': ['mae', 'mse'],
            'output_2': 'accuracy'
        }
    )
    

    # Train the model
    model.fit(X_train_fold, {'output_1': y_train_output1_fold, 'output_2': y_train_output2_fold},
          epochs=150, batch_size=128, verbose=0)

    
    # Evaluate the model on validation data for this fold
    results = model.evaluate(X_val_fold, {'output_1': y_val_output1_fold, 'output_2': y_val_output2_fold})
    
    mse_output1_list.append(results[2])  # Index 2 corresponds to MSE for 'output_1'
    accuracy_output2_list.append(results[5])  # Index 5 corresponds to accuracy for 'output_2'

```

```python
# Calculate the mean and standard deviation of evaluation metrics across all folds
mean_mse_output1 = np.mean(mse_output1_list)
mean_accuracy_output2 = np.mean(accuracy_output2_list)

std_mse_output1 = np.std(mse_output1_list)
std_accuracy_output2 = np.std(accuracy_output2_list)

# For regression task (output_1) - Calculate R-squared (R^2) score
y_true_output1 = np.concatenate([y_train_output1, y_test_output1])
y_pred_output1 = model.predict(np.concatenate([X_train_normalized, X_test_normalized]))[0]
r2_output1 = r2_score(y_true_output1, y_pred_output1)

print("Mean MSE (Output 1):", mean_mse_output1)
print("Mean Accuracy (Output 2):", mean_accuracy_output2)

print("Standard Deviation MSE (Output 1):", std_mse_output1)
print("Standard Deviation Accuracy (Output 2):", std_accuracy_output2)

print("R^2 Score (Output 1 - Regression):", r2_output1)
```

```python
# 1. Compare Regression Output with Original Test Values

# Predict using the trained model on the test data for the regression output (output_1)
y_pred_output1_test = model.predict(X_test_normalized)[0]

# Reshape the predictions and original test values to match for comparison
y_pred_output1_test = y_pred_output1_test.flatten()

# Convert both arrays to the same data type (e.g., float64) for comparison
y_test_output1 = y_test_output1.astype(np.float64)
y_pred_output1_test = y_pred_output1_test.astype(np.float64)

# Stack the original test values and predicted values horizontally
comparison_output1 = np.column_stack((y_test_output1, y_pred_output1_test))

# Print the original test values and predicted values side by side
print("Original Test Values (Output 1) vs. Predicted Values (Output 1):")
print(comparison_output1)
```

```python
# 2. Generate Confusion Matrix for Classification

# Predict using the trained model on the test data for the classification output (output_2)
y_pred_output2_test = model.predict(X_test_normalized)[1]

# Convert the probability predictions to binary class labels (0 or 1)
y_pred_output2_test_binary = np.round(y_pred_output2_test).astype(int).flatten()

# Import the confusion_matrix function from sklearn.metrics
from sklearn.metrics import confusion_matrix

# Generate the confusion matrix
conf_matrix_output2 = confusion_matrix(y_test_output2, y_pred_output2_test_binary)

# Print the confusion matrix
print("Confusion Matrix (Output 2 - Classification):")
print(conf_matrix_output2)
```
