## Removing columns whose entire column is zero

```python
def all_entries_nonzero(column):
    return (column != 0).any()

columns_to_drop = dataset.columns[~dataset.apply(all_entries_nonzero)]
dataset_new = dataset.drop(columns_to_drop, axis=1)
```

```python
data = dataset_new['param__func__de.upb.lb-nginx.0.1__mem_max']
encoded_data = pd.get_dummies(data, columns=['param__func__de.upb.lb-nginx.0.1__mem_max'])
dataset_new = dataset_new.drop('param__func__de.upb.lb-nginx.0.1__mem_max', axis = 1)
dataset_new = pd.concat([dataset_new, encoded_data], axis=1)
dataset_new
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Separate the two output features
y_train_output1 = y_train[:, 0]  # First column for the output ranging from 0.1 to 1.0
y_train_output2 = y_train[:, -4]  # Second column for binary classification

y_train_output3 = y_train[:, -3]
y_train_output4 = y_train[:, -2]
y_train_output5 = y_train[:, -1]

y_test_output1 = y_test[:, 0]  # First column for the output ranging from 0.1 to 1.0
y_test_output2 = y_test[:, -4]  # Second column for binary classification

y_test_output3 = y_test[:, -3]
y_test_output4 = y_test[:, -2]
y_test_output5 = y_test[:, -1]
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize input features
scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)

```

## Building Model

```python
input_shape = (32,)
inputs = tf.keras.Input(shape=input_shape)
x = tf.keras.layers.Dense(units=64, activation='relu')(inputs)
x = tf.keras.layers.Dropout(0.2)(x)
x = tf.keras.layers.Dense(units=64, activation='relu')(x)
x = tf.keras.layers.Dense(units=32, activation='relu')(x)
output_1 = tf.keras.layers.Dense(units=1, activation='linear', name='output_1')(x)  # Output 1: Regression
output_2 = tf.keras.layers.Dense(units=1, activation='sigmoid', name='output_2')(x)  # Output 2: Classification

output_3 = tf.keras.layers.Dense(units=1, activation='sigmoid', name='output_3')(x)  # Output 2: Classification
output_4 = tf.keras.layers.Dense(units=1, activation='sigmoid', name='output_4')(x)  # Output 2: Classification
output_5 = tf.keras.layers.Dense(units=1, activation='sigmoid', name='output_5')(x)  # Output 2: Classification

model = tf.keras.Model(inputs=inputs, outputs=[output_1, output_2, output_3, output_4, output_5])

# Check the model output names
print("Model Output Names:", model.output_names)  
```

```python
model.compile(
    loss={
        'output_1': 'mean_squared_error',  # Regression loss function
        'output_2': 'binary_crossentropy',  # Categorical classification loss function
        'output_3': 'binary_crossentropy',
        'output_4': 'binary_crossentropy',
        'output_5': 'binary_crossentropy',
    },
    optimizer='adam',
    metrics={
        'output_1': ['mae', 'mse'],
        'output_2': 'accuracy',
        'output_3': 'accuracy',
        'output_4': 'accuracy',
        'output_5': 'accuracy',
    }
)

model.fit(X_train_normalized, {'output_1': y_train_output1, 'output_2': y_train_output2, 'output_3': y_train_output3, 'output_4': y_train_output4, 'output_5': y_train_output5},
          validation_data=(X_test_normalized, {'output_1': y_test_output1, 'output_2': y_test_output2, 'output_3': y_test_output3, 'output_4': y_test_output4, 'output_5': y_test_output5}),
          epochs=50, batch_size=32)

```

```python
# After the model is trained, you can make predictions
y_pred_output1, y_pred_output2, y_pred_output3, y_pred_output4, y_pred_output5 = model.predict(X_test_normalized)

# For output_1 (Regression)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae_output1 = mean_absolute_error(y_test_output1, y_pred_output1)
mse_output1 = mean_squared_error(y_test_output1, y_pred_output1)
r2_output1 = r2_score(y_test_output1, y_pred_output1)

print("Output 1 - Mean Absolute Error:", mae_output1)
print("Output 1 - Mean Squared Error:", mse_output1)
print("Output 1 - R^2 Score:", r2_output1)

# For output_2 to output_5 (Classification)
from sklearn.metrics import accuracy_score, confusion_matrix

for i in range(4):
    cm = confusion_matrix(y_test_output2, np.round(y_pred_output2))
    accuracy = accuracy_score(y_test_output2, np.round(y_pred_output2))
    print(f"Confusion Matrix for Classification Output {i+2}:\n{cm}")
    print(f"Accuracy Score for Classification Output {i+2}: {accuracy}")


```

## K-fold Cross Validation

```python
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

num_folds = 5

# Initialize lists to store the evaluation metrics for each fold
mse_output1_list = []
accuracy_output2_list = []
accuracy_output3_list = []
accuracy_output4_list = []
accuracy_output5_list = []

# Create the KFold object
kfold = KFold(n_splits=num_folds, shuffle=True, random_state=42)


# Create the model
input_shape = (32,)
inputs = tf.keras.Input(shape=input_shape)
x = tf.keras.layers.Dense(units=256, activation='relu')(inputs)  # GELU activation

x = tf.keras.layers.Dense(units=256, activation='relu')(x)  # GELU activation

x = tf.keras.layers.Dense(units=256, activation='relu')(x)  # GELU activation

x = tf.keras.layers.Dense(units=256, activation='relu')(x)  # GELU activation

x = tf.keras.layers.Dense(units=128, activation='relu')(x)  # GELU activation

x = tf.keras.layers.Dense(units=128, activation='relu')(x)  # GELU activation

x = tf.keras.layers.Dense(units=128, activation='relu')(x)  # GELU activation
   

output_1 = tf.keras.layers.Dense(units=1, activation='linear', name='output_1')(x)  # Output 1: Regression
output_2 = tf.keras.layers.Dense(units=1, activation='sigmoid', name='output_2')(x)  # Output 2: Classification
output_3 = tf.keras.layers.Dense(units=1, activation='sigmoid', name='output_3')(x)
output_4 = tf.keras.layers.Dense(units=1, activation='sigmoid', name='output_4')(x)
output_5 = tf.keras.layers.Dense(units=1, activation='sigmoid', name='output_5')(x)

model = tf.keras.Model(inputs=inputs, outputs=[output_1, output_2, output_3, output_4, output_5])

# Perform 5-fold cross-validation
for fold, (train_indices, val_indices) in enumerate(kfold.split(X_train_normalized)):
    print(f"Fold {fold+1}/{num_folds}")
    
    # Split the data into training and validation sets for this fold
    X_train_fold = X_train_normalized[train_indices]
    X_val_fold = X_train_normalized[val_indices]
    y_train_output1_fold = y_train_output1[train_indices]
    y_train_output2_fold = y_train_output2[train_indices]
    y_train_output3_fold = y_train_output3[train_indices]
    y_train_output4_fold = y_train_output4[train_indices]
    y_train_output5_fold = y_train_output5[train_indices]
    
    y_val_output1_fold = y_train_output1[val_indices]
    y_val_output2_fold = y_train_output2[val_indices]
    y_val_output3_fold = y_train_output3[val_indices]
    y_val_output4_fold = y_train_output4[val_indices]
    y_val_output5_fold = y_train_output5[val_indices]
    
    
    # Compile the model
    model.compile(
        loss={
            'output_1': 'mean_squared_error',  # Regression loss function
            'output_2': 'binary_crossentropy',  # classification loss function
            'output_3': 'binary_crossentropy',
            'output_4': 'binary_crossentropy',
            'output_5': 'binary_crossentropy',
        },
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        metrics={
            'output_1': ['mae', 'mse'],
            'output_2': 'accuracy',
            'output_3': 'accuracy',
            'output_4': 'accuracy',
            'output_5': 'accuracy',
        }
    )
    

    # Train the model
    model.fit(X_train_fold, {'output_1': y_train_output1_fold, 'output_2': y_train_output2_fold, 'output_3': y_train_output3_fold, 'output_4': y_train_output4_fold, 'output_5': y_train_output5_fold},
          epochs=100, batch_size=32, verbose=0)

    
    # Evaluate the model on validation data for this fold
    results = model.evaluate(X_val_fold, {'output_1': y_val_output1_fold, 'output_2': y_val_output2_fold, 'output_3': y_val_output3_fold, 'output_4': y_val_output4_fold, 'output_5': y_val_output5_fold})
#     print("Results:", results)
    
    mse_output1_list.append(results[2])  # Index 2 corresponds to MSE for 'output_1'
    accuracy_output2_list.append(results[8])  # Index 5 corresponds to accuracy for 'output_2'
    accuracy_output3_list.append(results[9])  # Index 8 corresponds to accuracy for 'output_3'
    accuracy_output4_list.append(results[10])  # Index 11 corresponds to accuracy for 'output_4'
    accuracy_output5_list.append(results[11])  # Index 14 corresponds to accuracy for 'output_5'


```

```python
# Calculate the mean and standard deviation of evaluation metrics across all folds
mean_mse_output1 = np.mean(mse_output1_list)
mean_accuracy_output2 = np.mean(accuracy_output2_list)
mean_accuracy_output3 = np.mean(accuracy_output3_list)
mean_accuracy_output4 = np.mean(accuracy_output4_list)
mean_accuracy_output5 = np.mean(accuracy_output5_list)


std_mse_output1 = np.std(mse_output1_list)
std_accuracy_output2 = np.std(accuracy_output2_list)
std_accuracy_output3 = np.std(accuracy_output3_list)
std_accuracy_output4 = np.std(accuracy_output4_list)
std_accuracy_output5 = np.std(accuracy_output5_list)

# For regression task (output_1) - Calculate R-squared (R^2) score
y_true_output1 = np.concatenate([y_train_output1, y_test_output1])
y_pred_output1 = model.predict(np.concatenate([X_train_normalized, X_test_normalized]))[0]
r2_output1 = r2_score(y_true_output1, y_pred_output1)

print("Mean MSE (Output 1):", mean_mse_output1)
print("Mean Accuracy (Output 2):", mean_accuracy_output2)
print("Mean Accuracy (Output 3):", mean_accuracy_output3)
print("Mean Accuracy (Output 4):", mean_accuracy_output4)
print("Mean Accuracy (Output 5):", mean_accuracy_output5)

print("Standard Deviation MSE (Output 1):", std_mse_output1)
print("Standard Deviation Accuracy (Output 2):", std_accuracy_output2)
print("Standard Deviation Accuracy (Output 3):", std_accuracy_output3)
print("Standard Deviation Accuracy (Output 4):", std_accuracy_output4)
print("Standard Deviation Accuracy (Output 5):", std_accuracy_output5)

print("R^2 Score (Output 1 - Regression):", r2_output1)
```
